```python
xtrain = train.drop(['Activity', 'subject'], axis =1)
ytrain = train['Activity']

xtest = test.drop(['Activity', 'subject'], axis =1)
ytest = test['Activity']
```

```python
print("Training data size: ", xtrain.shape)
print("Testing data size: ", xtest.shape)
```

```python
parameters = {'max_iter': [100,200,500]}
lr_classifier = LogisticRegression()
lr_classifier_rs = RandomizedSearchCV(lr_classifier, cv= 5, random_state= 42, param_distributions = parameters)

# we use this randomisedcv as it breaks our data into batches (i.e cv=5 means 5 batches) 
#and train the model on those batches and gives us the cummulative result 

#we also use hyperparameters which will test our model on different parameters our lr_classifier_rs will run 15 times
#(i.e cv=5 * parameters and give us the best result from each parameters)


lr_classifier_rs.fit(xtrain,ytrain)
y_pred_lr = lr_classifier_rs.predict(xtest) # predict our data
```

```python
lr_accuracy = accuracy_score(y_true = ytest, y_pred = y_pred_lr)
print("Accuracy using Logistic Regression model is : ", lr_accuracy)
```

```python
def plot_confusion_matrix(cm,labels):
    fig, ax = plt.subplots(figsize=(12,8)) # for plotting confusion matrix as image
    im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    ax.figure.colorbar(im, ax=ax)
    ax.set(xticks=np.arange(cm.shape[1]),
    yticks=np.arange(cm.shape[0]),
    xticklabels=labels, yticklabels=labels,
    ylabel='True label',
    xlabel='Predicted label')
    plt.xticks(rotation = 90)
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, int(cm[i, j]),ha="center", va="center",color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
```

```python
cm = confusion_matrix(ytest.values,y_pred_lr)
cm
```

```python
cm = confusion_matrix(ytest.values,y_pred_lr)
plot_confusion_matrix(cm, np.unique(y_pred_lr))
```

```python
def get_best_randomsearch_results(model):
    print("Best estimator : ", model.best_estimator_)
    print("Best set of parameters : ", model.best_params_)
    #print("Best score : ", model.best_score_)
```

```python
get_best_randomsearch_results(lr_classifier_rs)
```
